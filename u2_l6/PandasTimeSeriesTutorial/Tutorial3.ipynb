```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from datetime import datetime
# pd.set_printoptions(notebook_repr_html=False)
```

```python
rng = pd.date_range('7/17/2012', periods=100000, freq='t') # Changed from datetime.now() to 2012-07-17 10:45:26.865218 then current form
rng
```

```python
ts = Series(np.random.randn(100000), index=rng)
```

```python
ts[ts.index[:15]]
```

```python
ts.ix[ts.index[:15]]
```

```python
ts.reindex(ts.index[:15])
```

```python
ts[pd.date_range(ts.index[0], ts.index[-1], freq='30t')] # Frequency of 30 min
```

```python
# Computing the mean value is resampling
# Downsampling: High freq to low
# Upsampling: low to high
ts.resample('30t') # Default aggregation method is mean 
```

```python
ts.head()
```

```python
ts.resample('30t', how='count') # Instead of mean, computes number of values that fall into each bin
```

```python
ts.resample('30t', how='count', closed='left') 
```
